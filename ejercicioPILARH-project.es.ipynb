#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame desde el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Supongamos que el subconjunto del Ejercicio 14 es un filtro específico, por ejemplo, propiedades en "arroyomolinos (Madrid)"
# Aquí, para generalizar, vamos a usar todo el dataset
subconjunto = ds.dropna(subset=['price', 'level5'])

# Normalizar la variable de precios para cada población
subconjunto['price_normalized'] = subconjunto.groupby('level5')['price'].transform(lambda x: (x - x.mean()) / x.std())

# Filtrar las poblaciones de interés
poblaciones = ['arroyomolinos (Madrid)', 'valdemorillo', 'galapagar', 'otra_poblacion']  # Reemplaza 'otra_poblacion' con una población real
subconjunto_filtrado = subconjunto[subconjunto['level5'].isin(poblaciones)]

# Crear el gráfico de histogramas
plt.figure(figsize=(6, 3))
for poblacion in poblaciones:
    data = subconjunto_filtrado[subconjunto_filtrado['level5'] == poblacion]['price_normalized']
    plt.hist(data, bins=20, alpha=0.5, label=poblacion)

plt.title('Histogramas Normalizados de Precios por Población')
plt.xlabel('Precio Normalizado (EUR)')
plt.ylabel('Frecuencia')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Conclusión
print("Conclusión: Los histogramas muestran la distribución normalizada de los precios por población. Podemos observar que cada población tiene una distribución diferente de precios, lo que indica que hay variabilidad en los precios de las casas entre las diferentes poblaciones.")
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# TODO
import pandas as pd

# Cargar el DataFrame desde el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Supongamos que el subconjunto del Ejercicio 14 es un filtro específico, por ejemplo, propiedades en "arroyomolinos (Madrid)"
# Aquí, para generalizar, vamos a usar todo el dataset
subconjunto = ds.dropna(subset=['price', 'surface', 'level5'])

# Crear una nueva columna 'pps' (precio por metro cuadrado)
subconjunto['pps'] = subconjunto['price'] / subconjunto['surface']

# Filtrar los municipios de interés
municipios = ['Getafe', 'Alcorcón']
subconjunto_filtrado = subconjunto[subconjunto['level5'].isin(municipios)]

# Calcular la media de 'pps' en ambos municipios
media_pps_getafe = subconjunto_filtrado[subconjunto_filtrado['level5'] == 'Getafe']['pps'].mean()
media_pps_alcorcon = subconjunto_filtrado[subconjunto_filtrado['level5'] == 'Alcorcón']['pps'].mean()

# Imprimir las medias de 'pps'
print(f"La media de precio por metro cuadrado en Getafe es de {media_pps_getafe:.2f} EUR/m²")
print(f"La media de precio por metro cuadrado en Alcorcón es de {media_pps_alcorcon:.2f} EUR/m²")

# Conclusión
print("Conclusión: Comparando los precios por metro cuadrado entre Getafe y Alcorcón, podemos observar que hay diferencias en los valores medios. Esto puede indicar variaciones en el mercado inmobiliario entre estos dos municipios.")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame desde el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Supongamos que el subconjunto del Ejercicio 14 es un filtro específico, por ejemplo, propiedades en "arroyomolinos (Madrid)"
# Aquí, para generalizar, vamos a usar todo el dataset
subconjunto = ds.dropna(subset=['price', 'surface', 'level5'])

# Filtrar las poblaciones de interés
poblaciones = ['Getafe', 'Alcorcón', 'arroyomolinos (Madrid)', 'valdemorillo']
subconjunto_filtrado = subconjunto[subconjunto['level5'].isin(poblaciones)]

# Crear subplots para cada población
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Lista de subplots
subplots = [(0, 0), (0, 1), (1, 0), (1, 1)]

# Crear un diagrama de dispersión para cada población
for i, poblacion in enumerate(poblaciones):
    data = subconjunto_filtrado[subconjunto_filtrado['level5'] == poblacion]
    row, col = subplots[i]
    axs[row, col].scatter(data['surface'], data['price'], alpha=0.5)
    axs[row, col].set_title(f'{poblacion}')
    axs[row, col].set_xlabel('Superficie (m²)')
    axs[row, col].set_ylabel('Precio (EUR)')
    axs[row, col].grid(True)

# Ajustar el diseño
plt.tight_layout()
plt.show()

# Conclusión
print("Conclusión: Los diagramas de dispersión muestran la relación entre la superficie y el precio de las casas en diferentes poblaciones. Podemos observar que cada población tiene una distribución diferente de precios y superficies, lo que indica variabilidad en el mercado inmobiliario entre estas poblaciones.")
```

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame desde el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Supongamos que el subconjunto del Ejercicio 14 es un filtro específico, por ejemplo, propiedades en "arroyomolinos (Madrid)"
# Aquí, para generalizar, vamos a usar todo el dataset
subconjunto = ds.dropna(subset=['latitude', 'longitude', 'level5'])

# Filtrar las poblaciones del cinturón sur de Madrid
poblaciones_sur = ['Getafe', 'Alcorcón', 'Leganés', 'Fuenlabrada', 'Móstoles']
subconjunto_filtrado = subconjunto[subconjunto['level5'].isin(poblaciones_sur)]

# Crear un diccionario de coordenadas por población
coordenadas = {}
for poblacion in poblaciones_sur:
    data = subconjunto_filtrado[subconjunto_filtrado['level5'] == poblacion]
    coordenadas[poblacion] = list(zip(data['latitude'], data['longitude']))

# Trazar las coordenadas en un mapa
plt.figure(figsize=(10, 8))
colors = ['red', 'blue', 'green', 'purple', 'orange']
for i, poblacion in enumerate(poblaciones_sur):
    latitudes, longitudes = zip(*coordenadas[poblacion])
    plt.scatter(longitudes, latitudes, c=colors[i], label=poblacion, alpha=0.5)

plt.title('Coordenadas del Cinturón Sur de Madrid por Población')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Conclusión
print("Conclusión: El trazado de las coordenadas muestra la distribución geográfica de las propiedades en el cinturón sur de Madrid. Cada población está representada por un color diferente, lo que permite visualizar la concentración de propiedades en cada área.")
```
