#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

```python
# TODO
from sklearn.preprocessing import MinMaxScaler

# Normaliza la variable de precios para cada población

# Crear un scaler
scaler = MinMaxScaler()

# Normalizar los precios para cada población
cinturon_sur['precio_normalizado'] = scaler.fit_transform(cinturon_sur[['precio']])

# Trazar los histogramas normalizados
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
plt.figure(figsize=(10, 6))

for poblacion in poblaciones:
    subset = cinturon_sur[cinturon_sur['level5'] == poblacion]
    plt.hist(subset['precio_normalizado'], bins=20, alpha=0.5, label=poblacion)

plt.title('Histogramas Normalizados de Precios en el Cinturón Sur de Madrid')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(loc='upper right')
plt.show()
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# TODO
# Filtra el DataFrame original para obtener solo las filas de las poblaciones del "cinturón sur" de Madrid
cinturon_sur = ds[ds['level5'].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])]

# Crea una nueva columna llamada `pps` (precio por metro cuadrado)
cinturon_sur['pps'] = cinturon_sur['precio'] / cinturon_sur['superficie']

# Filtra el dataset para obtener solo las filas de "Getafe" y "Alcorcón"
getafe = cinturon_sur[cinturon_sur['level5'] == "Getafe"]
alcorcon = cinturon_sur[cinturon_sur['level5'] == "Alcorcón"]

# Calcula la media de los precios por metro cuadrado
media_pps_getafe = getafe['pps'].mean()
media_pps_alcorcon = alcorcon['pps'].mean()

# Imprime los valores obtenidos
print(f"La media de precio por metro cuadrado en Getafe es de {media_pps_getafe} USD/m2")
print(f"La media de precio por metro cuadrado en Alcorcón es de {media_pps_alcorcon} USD/m2")

# Compara si las medias son iguales
son_iguales_pps = media_pps_getafe == media_pps_alcorcon
print(f"¿Son los precios promedios por metro cuadrado de Getafe y Alcorcón los mismos? {son_iguales_pps}")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
# TODO
import matplotlib.pyplot as plt

# Poblaciones a analizar
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Crear subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Iterar sobre las poblaciones y crear un scatter plot para cada una
for i, poblacion in enumerate(poblaciones):
    subset = cinturon_sur[cinturon_sur['level5'] == poblacion]
    ax = axs[i // 2, i % 2]
    ax.scatter(subset['superficie'], subset['precio'], alpha=0.5)
    ax.set_title(f'Relación entre Superficie y Precio en {poblacion}')
    ax.set_xlabel('Superficie (metros cuadrados)')
    ax.set_ylabel('Precio (USD)')

plt.tight_layout()
plt.show()
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:

from ipyleaflet import Map, Marker, MarkerCluster, basemaps
from ipywidgets import HTML

# Mapa centrado en Madrid
map = Map(center=(40.4168, -3.7038), zoom=10, basemap=basemaps.Stamen.Terrain)

# Diccionario de colores para cada población
colores = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "purple"
}

# Crear un cluster de marcadores
marker_cluster = MarkerCluster()

# Iterar sobre las poblaciones y agregar marcadores al mapa
for poblacion in poblaciones:
    subset = cinturon_sur[cinturon_sur['level5'] == poblacion]
    for _, row in subset.iterrows():
        marker = Marker(location=(row['latitud'], row['longitud']), draggable=False, icon_color=colores[poblacion])
        marker_cluster.add_layer(marker)

# Agregar el cluster de marcadores al mapa
map.add_layer(marker_cluster)

# Mostrar el mapa
map

```
